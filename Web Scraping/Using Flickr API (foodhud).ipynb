# Mining photos from Flickr

### Usage:

Need to get api_key and api_secret from Flickr account. Need to set "results_folder" as well. Then just change text='' part of "photos" variable.

### Notes:

I'm not sure how many images it's trying to download (seems like should be 50 but not 100% sure).

"url_o" means get image in original size. Set to "url_c" for "medium" sized image.

```python
import flickrapi
import urllib

api_key = u'xxxxxxxxxxxxxxxxxx'
api_secret = u'xxxxxxxxxxxxxxxx'


flickr = flickrapi.FlickrAPI(api_key, api_secret)

photos = flickr.walk(text='apple pie',
        sort='relevance',
        extras='url_o',
        per_page=50)                                    
#         min_taken_date='2008-08-20',
#         max_taken_date='2008-08-30')

count = 0
results_folder = "/path/to/file/"
for photo in photos:
    try:
        url=photo.get('url_o')
        print(url)
        count += 1
        urllib.request.urlretrieve(url,  results_folder + str(count) +".jpg")
    except Exception as e:
        print(e, 'Download failure')
    
```
